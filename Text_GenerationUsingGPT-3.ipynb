```python
# Load tokenizer and generator
tokenizer = AutoTokenizer.from_pretrained("EleutherAI/gpt-neo-1.3B")
generator= pipeline ('text-generation',
                     model='EleutherAI/gpt-neo-1.3B')
```

```python
# Data preprocessing function
def preprocess_text(text):
    try:
        text = text.lower()  # Convert to lowercase
        text = re.sub(r'[^a-zA-Z0-9\s]', '', text)  # Remove punctuation
        sentences = sent_tokenize(text)  # Sentence tokenization
        return " ".join(sentences)  # Combine sentences into a single string
    except Exception as e:
        print(f"Error preprocessing text: {e}")
        return ""
```

```python
# Text generation function
def generate_text(prompt, max_length=200, temperature=0.7, top_k=50, top_p=0.9):
    output = generator(
        prompt,
        max_length=max_length,
        temperature=temperature,
        top_k=top_k,
        top_p=top_p,
        num_return_sequences=1,
        pad_token_id=tokenizer.eos_token_id,
        no_repeat_ngram_size=2
    )
    return output[0]['generated_text']
```

```python
# Example usage
prompt = "In a distant galaxy, a brave warrior"
preprocessed_prompt = preprocess_text(prompt)
story = generate_text(preprocessed_prompt)

print("Generated text:")
print(story)
```
